 Load the Excel file robustly

```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path.cwd().parent / "data" / "raw"

# pick the first .xls or .xlsx file in data/raw
excel_files = list(RAW_DIR.glob("*.xls")) + list(RAW_DIR.glob("*.xlsx"))
print("Excel files found:", excel_files)

path = excel_files[0]  # assumes you have one file there
df = pd.read_excel(path, header=1, engine="xlrd")  # .xls needs xlrd

df.head()
```

```python
df = df.rename(columns={"default payment next month": "default"})
df.columns = [c.strip().lower().replace(" ", "_").replace(".", "_") for c in df.columns]
df.head()
```

Cleaning up dirty categories from the Data Set. Targeting Education and Marriage

```python
df["education"] = df["education"].replace({0: 4, 5: 4, 6: 4})  # 4 = other
df["marriage"]  = df["marriage"].replace({0: 3})             # 3 = other
```

Saving a clean copy of the data

```python
processed_path = Path.cwd().parent / "data" / "processed" / "credit_default_clean.csv"
processed_path.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(processed_path, index=False)
print("Saved to:", processed_path)
```
